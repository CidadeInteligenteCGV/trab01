# Controle de Frequência de Aulas

# Relatório 01

### Relatório da frequência geral dos alunos por disciplina, incluindo as seguintes informações: campus onde a disciplina é ofertada, curso a qual a disciplina pertence, código da disciplina, nome da disciplina, horas disciplina, nome do professor, nome do aluno, matrícula do aluno e frequência do aluno na disciplina. 

Código para obtenção do resultado:

```python
relatorio_01 = pd.read_sql_query("""
SELECT C.nome_disciplina AS Discilpina, C.cod_disciplina, C.horas_disciplina AS Carga_Horaria,  A.campus, B.nome_curso AS Curso, E.nome_professor AS Professor, G.nome_aluno AS Aluno, G.cod_matricula AS Matricula, D.data_aula AS Data_da_aula, D.presenca
FROM Campus A 
JOIN Curso B ON B.FK_Campus_cod_campus = A.cod_campus
JOIN Disciplina C  ON C.FK_Curso_cod_curso = B.cod_curso
JOIN Historico_Aula D  ON D.FK_Disciplina_cod_disciplina = C.cod_disciplina
JOIN Professor E ON E.cod_servidor = D.FK_Professor_cod_servidor
JOIN Gera F  ON F.FK_Historico_Aula_id_historico_aula = D.id_historico_aula
JOIN Aluno G  ON G.cod_matricula = F.FK_Aluno_cod_matricula
ORDER BY C.nome_disciplina
                            """,conn)
```

```python
relatorio_01
```

## Relatório 04

## Relatório 05

### Relatório de todas as disciplinas ofertadas por campus, incluindo as seguintes informações: código da disciplina, nome da disciplina, horas disciplina, nome do professor, curso onde a disciplina está alocada, campus onde a disciplina é ofertada.

codigo para obtenção do resultado:

```python
relatorio_05 = pd.read_sql_query("""
SELECT C.cod_disciplina, C.nome_disciplina, C.horas_disciplina,  B.nome_curso , A.campus, E.nome_professor
FROM Campus A 
JOIN Curso B ON B.FK_Campus_cod_campus = A.cod_campus
JOIN Disciplina C ON C.FK_Curso_cod_curso = B.cod_curso
JOIN Historico_Aula D ON D.FK_Disciplina_cod_disciplina = C.cod_disciplina
JOIN Professor E ON E.cod_servidor = D.FK_Professor_cod_servidor
ORDER BY C.horas_disciplina
                            """,conn)
```
